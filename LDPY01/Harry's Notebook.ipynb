```python
x = 0
print("Global x is: ", x)

def egfunc1():
    x = 1
    print("Local x in egfunc1 is: ", x)
    def thirdfunc():
        x = 2
        print("Local x in egfunc2 is: ", x) 
        return x
    thirdfunc()
#    return x
    
def egfunc3():
    global x # let x refer to the global variable instance within this function from now on
    print("Global x from within egfunc3 is: ", x)
    x = 3 # changes the global reference
    print("Global x from within egfunc3 is now: ", x)


# call the functions
egfunc1()
print("Global x is still: ", x)
egfunc3()
print("Global x is now: ", x)

print(x)
```

```python
import subprocess

class Host:

        def __init__(self, ip_address):
                self.ip_address = ip_address

        def is_alive(self):
                process = subprocess.Popen('ping -c2 {0} 2>&1'.format(self.ip_address), stdout=subprocess.PIPE, shell=True)
                output = process.stdout.read().decode('UTF-8')
                self.output = output
                lines = output.split('\n')
                for line in lines:
                        if 'received' in line:
                                received_count = int(line.split()[3])
                                if received_count > 0:
                                        return True
                return False

my_server = Host('8.8.8.8')
print(my_server.ip_address)
my_server.is_alive()
print(my_server.is_alive())
```
